```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data
df = pd.read_csv(r'Desktop/oluaka/1- mental-illnesses-prevalence.csv')

# Aggregate global averages by year
global_trends = df.groupby('Year').mean(numeric_only=True)

# Plot
plt.figure(figsize=(12, 6))
for column in global_trends.columns:
    plt.plot(global_trends.index, global_trends[column], label=column)
plt.title('Global Trends in Mental Disorder Prevalence (1990-2019)')
plt.xlabel('Year')
plt.ylabel('Prevalence (%)')
plt.legend()
plt.grid(True)
plt.show()
```
